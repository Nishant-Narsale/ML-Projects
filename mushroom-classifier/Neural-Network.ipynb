```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
```

```python
df = pd.read_csv('mushrooms.csv')
df
```

```python
df.groupby('class').count()
```

```python
df['class'].value_counts()
```

```python
df.columns
```

```python
lst = [i for i in range(50)]
for column_name in df.columns:
    indexes = df[column_name].value_counts().index
    df[column_name].replace(indexes, lst[:len(indexes)], inplace=True)
```

```python
df
```

```python
sns.barplot(x=df['cap-shape'],y=df['cap-surface'],hue=df['class'])
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = Sequential(
    [
        Dense(22, activation="relu", name="layer1"),
        Dense(44, activation="relu", name="layer2"),
        Dense(1, name="layer3"),
    ]
)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['binary_accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size = 128, epochs = 50)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
n_predictions = []
for i in predictions:
    if i>0:
        n_predictions.append(1)
    else:
        n_predictions.append(0)
n_predictions
```

```python
accuracy_score(y_test, n_predictions)*100
```
